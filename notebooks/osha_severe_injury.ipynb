```python
df = df.rename(columns={
    "Primary NAICS": "Primary_NAICS",
    "Part of Body": "Part_of_Body",
    "Part of Body Title": "Part_of_Body_Title",
    "Final Narrative": "Final_Narrative",
    "Secondary Source": "Secondary_Source",
    "Secondary Source Title": "Secondary_Source_Title"
})
```

```python
df.columns.tolist()
df.dtypes
```

```python
df["EventDate"] = pd.to_datetime(df["EventDate"])
df["Zip"] = df["Zip"].astype("Int64")
df["Amputation"] = df["Amputation"].fillna(0).astype("Int64")
df["Loss of Eye"] = df["Loss of Eye"].fillna(0).astype("Int64")
```
