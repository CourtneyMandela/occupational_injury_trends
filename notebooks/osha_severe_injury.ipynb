## Data Limitations and Ethical Considerations

### Coverage Limitations: Federal vs. State Plan Jurisdictions

This dataset contains severe injury reports submitted to **Federal OSHA only**. According to OSHA's documentation, the data "represent incidents under federal OSHA jurisdiction only. They do not include incidents in states covered by OSHA-approved State Plans" ([source](https://www.osha.gov/severeinjury)).

Twenty-two states (plus Puerto Rico) operate their own OSHA programs covering private-sector workers: Alaska, Arizona, California, Hawaii, Indiana, Iowa, Kentucky, Maryland, Michigan, Minnesota, Nevada, New Mexico, North Carolina, Oregon, South Carolina, Tennessee, Utah, Vermont, Virginia, Washington, and Wyoming. In these states, only employers under narrow federal jurisdiction (federal agencies, USPS, maritime workers) appear in this dataset.

**Implication:** State-level comparisons are not valid. California appears to have fewer injuries than far smaller states; not because California necessarily is safer, but because the vast majority of California employers report to Cal/OSHA, not federal OSHA. Any geographic analysis must either exclude state-plan states or avoid state-level claims entirely.

### Reporting Threshold Bias

The OSHA Severe Injury Reports (SIR) dataset is based on employer reports required under 29 CFR 1904.39 and includes incidents involving only three severe outcomes: in-patient hospitalization, amputation, and loss of an eye (fatalities are also reportable to OSHA under 1904.39, but are not part of the SIR dataset).

Because the dataset is limited to these reporting triggers, it does not capture many workplace harms that fall outside them, including:

- Injuries treated and released from emergency rooms (i.e., cases without in-patient admission)
- Most chronic or cumulative occupational conditions (e.g., hearing loss, repetitive strain, respiratory illness), unless they result in a reportable outcome
- Near-misses and minor injuries that do not meet OSHA’s severe injury reporting thresholds

Accordingly, the SIR dataset should be interpreted as a record of incidents meeting OSHA’s severe injury reporting criteria, not as a comprehensive measure of workplace injury and illness.


**Implication:** This data captures catastrophic outcomes, not overall workplace safety. An employer with poor day-to-day safety practices could appear "safe" if they happen to avoid these specific reportable thresholds.

### Counts vs. Rates

Raw injury counts do not account for workforce size. Large employers (Walmart, USPS, Amazon) appear frequently not necessarily because they are more dangerous, but because they employ more workers. Without employee headcount data, injury *rates* cannot be calculated.

**Implication:** Employer-level comparisons based on raw counts would be misleading. This analysis will avoid ranking employers by injury count.

### Employer Name Inconsistency

Employer names are not standardized. The same company may appear under multiple variations (e.g., "UNITED PARCEL SERVICE", "UPS", "UPS INC"), and companies may report under subsidiary, LLC, or DBA names.

**Implication:** Aggregating by employer requires careful cleaning and may still contain errors.

### Temporal Confounders

The 2020–2021 drop in reported injuries likely reflects reduced economic activity during the COVID-19 pandemic, not improved safety. Fewer workers on job sites means fewer opportunities for injury.

**Implication:** Year-over-year trends must account for this confounder. The 2020 dip should not be interpreted as a safety improvement.

### Demographic Blindness

The dataset does not include worker-level demographics such as age, race, gender, job tenure, or language proficiency. Prior research shows that injury risk is strongly patterned by these factors. New workers face substantially elevated risk—workers in their first month on the job experience roughly four times the lost-time injury rate of workers with more than one year of tenure, independent of age ([Institute for Work & Health](https://www.iwh.on.ca/plain-language-summaries/newness-and-risk-of-occupational-injury)). Temporary workers exhibit substantially higher *crude* injury rates than permanent workers, although this difference largely disappears after adjusting for occupation and job tenure ([Benavides et al., 2006](https://pmc.ncbi.nlm.nih.gov/articles/PMC2078100/)). In addition, workers with limited English proficiency are disproportionately employed in physically demanding and higher-risk occupations ([NCBI, 2017](https://www.ncbi.nlm.nih.gov/books/NBK525209/)).

**Implication:** Because these demographic and structural risk factors are not observable in the dataset, this analysis cannot determine whether reported injuries disproportionately affect vulnerable worker populations.


### Analysis Direction

Given these limitations, this project will focus on **injury characteristics** - body parts, injury types, and event mechanisms - rather than geographic or employer-level comparisons. The industry (NAICS) and injury classification data remain valid across the dataset, even though geographic representativeness is compromised.

### Body Part Analysis: Most Frequently Injured Body Parts

**Question:** Where on the body do severe workplace injuries occur most frequently?

**Chart type:** Horizontal bar chart. Bar charts are appropriate for comparing counts across discrete categories, and horizontal orientation keeps labels readable.

**Key finding:** Fingers and fingertips dominate severe injury reports, accounting for three of the top four specific body part categories. This pattern aligns with OSHA’s national “Seven-Year Lookback” analysis, which found upper extremities accounted for 40% of all severe injury reports from 2015–2021, with fingertips alone representing 29% of upper extremity injuries ([OSHA Seven-Year Lookback](https://www.osha.gov/sites/default/files/severe-injury-report-2015to2021.pdf)). Finger injuries - especially fingertips - also commonly involve contact with machinery and are frequently associated with amputation events.

**Design choices:**
- Bars sorted by frequency (highest at top) to emphasize ranking
- Single neutral color; no color encoding needed for this comparison
- X-axis starts at zero to avoid exaggerating differences
- Removed unnecessary chart elements (top/right spines) to reduce visual clutter
- Clear axis labels and title specifying data source and date range

**Limitations:** Some OSHA coding categories are non-specific (e.g., "Multiple body parts, n.e.c.", "Nonclassifiable"). These are retained to accurately reflect how injuries are recorded, but they reduce anatomical precision for those cases.


```python
# what body parts are most subject to severe injury? 
# will use a bar chart as they are good for comparing numerical values and showing rankings 

#get top 10 most frequently injured body parts
top_body_parts = df["Part_of_Body_Title"].value_counts().head(10)

#plot out the results 
fig, ax = plt.subplots(figsize=(10, 6))
top_body_parts.sort_values().plot(kind="barh", ax=ax, color="#1f77b4")

# label chart and axes
ax.set_xlabel("Number of Severe Injury Reports")
ax.set_ylabel("Body Part")
ax.set_title("Most Frequently Injured Body Parts in OSHA Severe Injury Reports (2015–2024)")

#Neutral colors used, clear easily understandable labelling, ranked from highest to lowest frequnecy, x-axis starts at 0

# Remove top and right spines for cleaner look
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.tight_layout()
plt.show()
```

#### Interpreting Non-Specific Body-Part Categories (OIICS)

This dataset uses the **BLS Occupational Injury and Illness Classification System (OIICS)** for “Part of Body Affected.” Some categories are intentionally broad and should be interpreted as *coding buckets*, not precise anatomy.

- **Body systems**: used when the condition affects an entire body system rather than a single anatomical part (OIICS Part of Body **Division 6**).  
- **Multiple body parts**: used when injuries involve body parts from **two or more divisions** (OIICS Part of Body **Division 8**).  
- **Other body parts**: includes prosthetic/orthopedic devices and body parts not classified elsewhere (OIICS Part of Body **Division 9**).  
- **n.e.c.**: means **“not elsewhere classified”** - a residual category used when the detail is known but doesn’t fit a more specific listed code.  
- **Nonclassifiable**: used when available information is insufficient to assign a more specific body-part code.

**Documentation:**
- BLS OIICS Manual (Part of Body code titles / division definitions): https://www.bls.gov/iif/definitions/oiics-manual-2010-section-3-2.pdf  
- BLS OIICS Manual landing page: https://www.bls.gov/iif/definitions/occupational-injuries-and-illnesses-classification-manual.htm  
- Example definition of *n.e.c.* (“not elsewhere classified”) in injury classification reporting: https://www.dir.ca.gov/DOSH/cfoi/CFOI_2023/cfoi2023-A-7.pdf


### Event Analysis: Most Common Injury Event Types

**Question:** What event or exposure most commonly leads to a severe injury report?

**Chart type:** Horizontal bar chart. This format compares counts across discrete event categories, and the horizontal orientation keeps long labels readable.

**Key finding:** The event distribution is strongly top-heavy. Although the dataset contains **633** distinct event categories, the **top 10** account for **~39%** of all severe injury reports. For comparison, if reports were evenly distributed across categories, the top 10 would represent only ~1.6% (10/633). The highest-frequency events are dominated by machinery-related mechanisms (caught in running equipment; compressed/pinched; struck against moving parts) and falls to lower levels, indicating that a small set of recurring event patterns contributes a substantial share of severe injury reports.

**Design choices:**
- Top 10 categories to keep the chart readable while highlighting the highest-frequency mechanisms
- Bars sorted by frequency (highest at top) to emphasize ranking
- Single neutral color; no color encoding needed for counts
- X-axis starts at zero to avoid exaggerating differences
- Value labels added to support quick comparison

**Limitations:** Event categories follow standardized coding and may be broad in some cases (e.g., “unspecified” falls). Narrative text can provide additional context but is not consistently structured for analysis.

```python
# Count how many reports fall into each event category (human-readable labels)
event_counts = df["EventTitle"].value_counts()

# For the chart, limit to the Top N categories for readability
top_n = 10
top_events = event_counts.head(top_n).sort_values()  # sort so barh displays in rank order

fig, ax = plt.subplots(figsize=(10, 6))
ax.barh(top_events.index, top_events.values)

ax.set_title(f"Top {top_n} Injury Event Types in OSHA Severe Injury Reports")
ax.set_xlabel("Number of reports")
ax.set_ylabel("Event type")

# Improve readability / reduce clutter
ax.grid(axis="x", linestyle="--", alpha=0.3)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

# Add value labels so readers don't have to estimate from the axis
ax.bar_label(ax.containers[0], padding=3, fontsize=9)
ax.tick_params(axis="y", labelsize=9)

plt.tight_layout()
plt.show()
```

```python
# Quantify concentration: what share of ALL reports are captured by the Top 10 categories?
top10_share = event_counts.head(10).sum() / event_counts.sum()
top10_share
```

```python
# Context for concentration: how many distinct event categories exist in the full dataset?
df["EventTitle"].nunique()
```

#### Why I computed “Top 10” *and* concentration metrics

I first used `value_counts()` on `EventTitle` to measure how frequently each event mechanism appears in the dataset.  
Because there are **633** distinct event categories, plotting all of them would produce an unreadable chart, so I plotted only the **top 10** categories to show the most common mechanisms clearly.

However, a “Top 10” chart alone does not show whether the distribution is truly concentrated or just has a long list of categories. To quantify this, I calculated:

- **Top-10 share** = (sum of counts in the top 10 categories) / (sum of counts across all categories)

This value is **~0.392**, meaning the top 10 event types account for **~39%** of all reports. I also calculated `nunique()` (= **633**) to provide context: capturing ~39% of reports with only 10 categories indicates a strongly top-heavy distribution, with a long tail of many low-frequency event types.

**About `EventTitle`:** This analysis uses `EventTitle`, which is the human-readable label for the dataset’s OIICS-coded “Event or Exposure” field (`Event`). OSHA codes Severe Injury Reports using the Occupational Injury and Illness Classification System (OIICS) before publishing the dataset. This means event categories come from a standardized taxonomy rather than being free-text descriptions. The free-text incident description is stored separately in `Final_Narrative`.
